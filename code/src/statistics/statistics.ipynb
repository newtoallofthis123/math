# Measure of central tendency

Measures of central tendency are measures that locate the center of a distribution of data. The mean, median, and mode are the three most commonly used measures of central tendency.

## Mean

```python
sample = [2, 4, 1, 6, 5, 3, 7, 8, 9]

mean = sum(sample) / len(sample)
print(mean)
```

### Weighted Mean

Weighted mean on the other hand doesn't give an equal weight to all the values. It is the sum of the product of the values and their weights divided by the sum of the weights.

```python
sample = [2, 4, 1, 6, 5, 3, 7, 8, 9]
weights = [i/sum(sample) for i in sample]

mean = sum([a*b for a,b in zip(sample, weights)]) / len(sample)
print(mean)
```

### Median

The median is the midpoint of a distribution of data. It is the value that divides the data into two equal parts. Half the values are less than or equal to the median and half the values are greater than or equal to the median. The median is the middle value when the data are arranged in order of increasing magnitude.

```python
sample = [2, 4, 1, 6, 5, 3, 7, 8, 9]

def median(sample):
    ordered = sorted(sample)
    n = len(sample)
    mid = int(n/2) - 1 if n % 2 == 0 else int(n/2)

    if n % 2 == 0:
        return (ordered[mid] + ordered[mid+1]) / 2
    else:
        return ordered[mid]
    
print(median(sample))
```

### Mode

Mode is the value that occurs most frequently in a data set. A data set can have more than one mode or no mode at all. A data set with two modes is called bimodal. A data set with more than two modes is called multimodal. A data set with no mode is called amodal.

```python
from collections import defaultdict

sample = [1, 3, 2, 5, 7, 0, 2, 3]


def mode(values):
    counts = defaultdict(lambda: 0)

    for s in values:
        counts[s] += 1

    max_count = max(counts.values())
    modes = [v for v in set(values) if counts[v] == max_count]
    return modes

print(mode(sample))
```

## Variance

Variance of a data set defines how much the data is spread out. It is the average of the squared differences from the mean.

```python
data = [0, 1, 5, 7, 9, 10, 14]

def variance_cal(data):
    mean = sum(data)/len(data)
    variance = sum([(x-mean)**2 for x in data]) / len(data)
    return variance

print(variance_cal(data))
```

## The Gaussian Distribution

The Gaussian distribution is a continuous function which approximates the exact binomial distribution of events.

If we plot a histogram of a random gaussian distribution, we get something like this

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Set the mean and standard deviation of the distribution
mean = 0
std_dev = 1

# Generate x values for the curve
x = np.linspace(-4, 4, 100)

# Calculate the corresponding y values using the normal distribution
y = norm.pdf(x, mean, std_dev)

# Plot the bell curve
plt.plot(x, y)

# Add labels and title
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.title('Bell Curve')

# Display the plot
plt.show()
```

```python
from scipy.stats import norm

mean = 64.43
std_dev = 2.86

x = norm.cdf(64.2, mean, std_dev)
print("P(X<=64.2) = ", x)
```

```python
import random
from scipy.stats import norm
from matplotlib import pyplot as plt

weights = []

for i in range(0, 1000):
    random_p = random.uniform(0.0, 1.0)
    random_weight = norm.ppf(random_p,  loc=64.43, scale=2.99)
    weights.append(random_weight)

weights = [round(w, 2) for w in weights]

plt.hist(weights, bins=20, edgecolor='black')
plt.xlabel('Weight')
plt.ylabel('Frequency')
plt.title('Distribution of Weights')
plt.show()
```

## Central Limit Theorem

The central limit theorem states that the sampling distribution of the mean of any independent, random variable will be normal or nearly normal, if the sample size is large enough.

```python
import matplotlib.pyplot as plt
import random

sample_size = 31
sample_count = 1000

x_values = [(sum([random.uniform(0.0, 1.0) for i in range(
    sample_size)]) / sample_size) for _ in range(sample_count)]

plt.hist(x_values, bins=20)
plt.xlabel('Mean')
plt.ylabel('Frequency')
plt.title('Distribution of Sample Means')
plt.show()
```
