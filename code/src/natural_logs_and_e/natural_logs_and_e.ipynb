```python
def derivative_cal(f, x, h):
    return (f(x + h) - f(x)) / h


def f(x):
    return x ** 2


print(derivative_cal(f, 2, 0.0001))
```

```python
from sympy import *

x, y = symbols('x y')

f = 2 * x ** 3 + 3 * y ** 3

d_x = diff(f, x).subs([(x, 2), (y, 3)])
d_y = diff(f, y).subs([(x, 2), (y, 3)])

print(d_x / d_y)
```

```python
from sympy import *
from sympy.plotting import plot3d

# Declare x and y to SymPy
x, y = symbols('x y')

# Now just use Python syntax to declare function
f = 2*x**3 + 3*y**3

# Calculate the partial derivatives for x and y
dx_f = diff(f, x)
dy_f = diff(f, y)

print(dx_f)  # prints 6*x**2
print(dy_f)  # prints 9*y**2

# plot the function
plot3d(f)
```
